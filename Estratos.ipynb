```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/lihkir/Uninorte/main/AppliedStatisticMS/DataVisualizationRPython/Lectures/Python/PythonDataSets/googleplaystore.csv")
```

```python

df = df.rename(columns={
    'App': 'app',
    'Category': 'category',
    'Rating': 'rating',
    'Reviews': 'reviews',
    'Size': 'size',
    'Installs': 'installs',
    'Type': 'type',
    'Price': 'price',
    'Content Rating': 'content_rating',
    'Genres': 'genre',
    'Last Updated': 'last_updated',
    'Current Ver': 'current_ver',
    'Android Ver': 'android_ver'
})
df = df.dropna()
```

```python
print(df.head)
```

Grafico de barras estratificado 

```python
# Tomar una muestra aleatoria de tus datos
df_sample = df.sample(n=1500, random_state=50)  

estratifiacion = alt.Chart(df_sample).mark_bar().encode(
    x=alt.X('count():Q', title='NÃºmero de Aplicaciones'),
    y=alt.Y('category:N', title='CategorÃ­a'),
    color='content_rating:N'
).properties(
    title='NÃºmero de Aplicaciones Estratificado por Content Rating'
)
estratifiacion.show()
# Mostrar el grÃ¡fico
```

En este grafico, vemos cada categoria del dataframe, como, por ejemplo, Art_design, Lifestyle, Parenting,
Personalization, Tools, Weather, etc. Para cada categoria vemos, de manera acumulada, el numero de aplicaciones. 
Ademas, vemos la distribucion de cada numero de aplicacion en las categorias por el rating del cotenido. 

```python
heatmap = alt.Chart(df_sample).mark_rect().encode(
    x=alt.X('category:N', title='CategorÃ­a'),
    y=alt.Y('rating:Q', title='Rating', bin=alt.Bin(maxbins=10)),
    color='count():Q'
).properties(
    width=600,
    height=400,
    title='NÃºmero de Aplicaciones Estratificadas por CategorÃ­a y Rangos de Rating'
)
heatmap.show()
```

Ahora tenemos tres variables, la cuales son: Categoria, el rating y la cantidad de records: el cual va de una escala gris a una azul.
Para valores como la categoria: Family, vemos que para valores mayores de 3.5 en rating aumenta la cantidad de records. Ademas, de que en la 
mayoria de categorias, se ve un aumento en la cantidad de records, despues de 3.5 en rating.
